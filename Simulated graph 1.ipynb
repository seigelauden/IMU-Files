```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random sequence centered around 0
np.random.seed(42)  # For reproducibility
acceleration = np.random.uniform(-500, 500, 500) * 0.00981  # Convert to m/s^2

# Adjust the time step
dt = 0.1  # Smaller time step

# First integration to get velocity
velocity = np.cumsum(acceleration) * dt

# Second integration to get position
position = np.cumsum(velocity) * dt

# Plot the graphs
plt.figure(figsize=(14, 10))

# Plot original acceleration
plt.subplot(3, 1, 1)
plt.plot(acceleration, label='Acceleration (m/s^2)')
plt.title('Acceleration')
plt.ylabel('Acceleration (m/s^2)')
plt.legend()

# Plot velocity (first integration)
plt.subplot(3, 1, 2)
plt.plot(velocity, label='Velocity (m/s)', color='orange')
plt.title('Velocity (First Integration)')
plt.ylabel('Velocity (m/s)')
plt.legend()

# Plot position (second integration)
plt.subplot(3, 1, 3)
plt.plot(position, label='Position (m)', color='green')
plt.title('Position (Second Integration)')
plt.ylabel('Position (m)')
plt.legend()

plt.xlabel('Time')
plt.tight_layout()
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import cumtrapz

# Generate random sequence centered around 0
np.random.seed(42)  # For reproducibility
acceleration = np.random.uniform(-500, 500, 500) * 0.00981  # Convert to m/s^2

# Define the time step
dt = 0.1  # Smaller time step
time = np.arange(0, len(acceleration) * dt, dt)

# First integration to get velocity
velocity = cumtrapz(acceleration, time, initial=0)

# Second integration to get position
position = cumtrapz(velocity, time, initial=0)

# Plot the graphs
plt.figure(figsize=(14, 10))

# Plot original acceleration
plt.subplot(3, 1, 1)
plt.plot(time, acceleration, label='Acceleration (m/s^2)')
plt.title('Acceleration')
plt.ylabel('Acceleration (m/s^2)')
plt.legend()

# Plot velocity (first integration)
plt.subplot(3, 1, 2)
plt.plot(time, velocity, label='Velocity (m/s)', color='orange')
plt.title('Velocity (First Integration)')
plt.ylabel('Velocity (m/s)')
plt.legend()

# Plot position (second integration)
plt.subplot(3, 1, 3)
plt.plot(time, position, label='Position (m)', color='green')
plt.title('Position (Second Integration)')
plt.ylabel('Position (m)')
plt.legend()

plt.xlabel('Time (s)')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random sequence centered around 0
np.random.seed(42)  # For reproducibility
acceleration = np.random.uniform(-50, 50, 300) * 0.0981  # Convert to m/s^2

# Define the time step
dt = 0.1  # Smaller time step
time = np.arange(0, len(acceleration) * dt, dt)

# First integration to get velocity
velocity = np.zeros_like(acceleration)
for i in range(1, len(acceleration)):
    velocity[i] = velocity[i-1] + np.trapz(acceleration[i-1:i+1], dx=dt)

# Second integration to get position
position = np.zeros_like(velocity)
for i in range(1, len(velocity)):
    position[i] = position[i-1] + np.trapz(velocity[i-1:i+1], dx=dt)

# Plot the graphs
plt.figure(figsize=(14, 10))

# Plot original acceleration
plt.subplot(3, 1, 1)
plt.plot(time, acceleration, label='Acceleration (m/s^2)')
plt.title('Acceleration')
plt.ylabel('Acceleration (m/s^2)')
plt.legend()

# Plot velocity (first integration)
plt.subplot(3, 1, 2)
plt.plot(time, velocity, label='Velocity (m/s)', color='orange')
plt.title('Velocity (First Integration)')
plt.ylabel('Velocity (m/s)')
plt.legend()

# Plot position (second integration)
plt.subplot(3, 1, 3)
plt.plot(time, position, label='Position (m)', color='green')
plt.title('Position (Second Integration)')
plt.ylabel('Position (m)')
plt.legend()

plt.xlabel('Time (s)')
plt.tight_layout()
plt.show()
```
