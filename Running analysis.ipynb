```python
import pandas as pd
accelo_path = 'accelerometer.csv'

column_names = ["Date", "HostTimestamp", "NodeName", "NodeTimestamp", "RawData", "X (mg)", "Y (mg)", "Z (mg)"]

acceloData = pd.read_csv(accelo_path, skiprows=4, names=column_names)
acceloData = acceloData[acceloData['Date'] != 'Date']
print(acceloData.head())
```

```python
import matplotlib.pyplot as plt

acceloData['X (mg)'] = pd.to_numeric(acceloData['X (mg)'])
acceloData['Y (mg)'] = pd.to_numeric(acceloData['Y (mg)'])
acceloData['Z (mg)'] = pd.to_numeric(acceloData['Z (mg)'])

acceloData['Date'] = pd.to_datetime(acceloData['Date'])
# Extract only the seconds part of the time for the x-axis
acceloData['Time'] = acceloData['Date'].dt.strftime('%H:%M:%S.%f')

plt.figure(figsize=(12, 6))

subset_data = acceloData.iloc[80:262]


plt.plot(subset_data['Date'], subset_data['Z (mg)'], label='Z (mg)')


plt.xlabel('Date')
plt.ylabel('Acceleration (mg)')
plt.title('Acceleration in Z Direction')
plt.legend()

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
